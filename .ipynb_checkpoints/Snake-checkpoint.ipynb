# Snake programozási feladat

### Feladat

* mozgó kígyó, mely kaját eszik, és ezzel növekszik
* ha a kígyó elhagyja a pályát, a szemközti oldal megfelelő pontjából folytatja a mozgását
* a játék addig tart, amíg a kigyó önmagával nem ütközik, vagy a teljes képernyőt betölti

### Technikai adatok

* pálya mérete paraméterezhető
* kígyó mozgása egyre gyorsul
* megevett kaja pontot ér


```python
palya=[]
meret=10
for i in range(1,meret*meret+1):
    list.append(palya,'x')

def Kiir(palya,meret):
    for i in range(0,meret):
        for j in range(0,meret):
            print(palya[i*meret+j],' ',end='')
        print('')
    print('')

import random
kigyo=random.randint(0,64)
palya[kigyo]='O'

Kiir(palya,meret)
import time
time.sleep(1)


for i in range(1,10):
    irany=random.randint(1,4)

    if irany is 1:
        palya[kigyo]='x'
        kigyo=kigyo-meret
        if kigyo < 0:
            kigyo=64-abs(kigyo)
        palya[kigyo]='O'

    if irany is 2:
        palya[kigyo]='x'
        kigyo=kigyo+1
        if kigyo > 0:
            kigyo=abs(kigyo)
        palya[kigyo]='O'

    if irany is 1:
        palya[kigyo]='x'
        kigyo=kigyo+meret
        if kigyo > 0:
            kigyo=abs(kigyo)
        palya[kigyo]='O'

    if irany is 1:
        palya[kigyo]='x'
        kigyo=kigyo-1
        if kigyo < 0:
            kigyo=64-abs(kigyo)
        palya[kigyo]='O'

    Kiir(palya,meret)
    
    time.sleep(1)
```
